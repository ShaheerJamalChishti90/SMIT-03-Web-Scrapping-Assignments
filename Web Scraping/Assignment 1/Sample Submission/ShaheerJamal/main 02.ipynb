```python
from bs4 import BeautifulSoup
import requests
import json
import re
```

```python
#Scrapping Religion and Spirituality Books

url = "https://www.thriftbooks.com/browse/#b.s=mostPopular-desc&b.p=1&b.pp=50&b.f.t%5B%5D=14859"
web = requests.get(url).content
soup = BeautifulSoup(web, "lxml")
```

```python
content = soup.find_all("script")
content
```

```python
string = content[12].string
```

```python
main_data = re.search(r'window\.searchStoreV2\s*=\s*(\{.*?\});', string, re.DOTALL)

booksdata_dic = {}

if main_data:
    works = main_data.group(1)
    works_json = json.loads(works)
    works = works_json.get("works")
    
# with open(r'Religion and Sprituallity.csv', 'a') as f:
#     f.write(f'title, condition, buy_price$, language\n')
    
for i in works:
    booksdata_dic.update(i)
    title = booksdata_dic['title']
    condition = booksdata_dic['buyNowCondition']   
    buy_price = booksdata_dic['buyNowPrice']
    language = booksdata_dic['language']
    
    
    with open(r'Religion and Sprituality.csv', 'a') as f:
        f.write(f'{title}, {condition}, {buy_price}$, {language}\n')
```
