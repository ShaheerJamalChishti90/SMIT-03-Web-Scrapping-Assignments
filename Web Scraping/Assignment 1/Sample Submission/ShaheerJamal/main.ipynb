importing all modules from the library

```python
from bs4 import BeautifulSoup
import requests
import re
import json
```

setting up everything 

```python
#scraping 8K(approx) history books
url = "https://www.thriftbooks.com/browse/?13526col#b.s=mostPopular-desc&b.p=1&b.pp=50&b.col&b.f.t%5B%5D=13526&b.list"
web = requests.get(url).content
soup = BeautifulSoup(web, 'lxml')
soup
```

Working on the main webpage

```python
content = soup.find_all('script')
content[12]
```

```python
data = soup.find_all("script")
string = data[12].string

data = []

match = re.search(r"window.searchStoreV2\s*=\s*(\{.*?});", string, re.DOTALL)
if match:
    works = match.group(1)
    works_json = json.loads(works)
    works = works_json.get("works")
   
with open("History_Books_Data.csv", "w") as f:
    f.write(f"Title, Language\n")

    
for i in works:
    title = i.get("title")
    language = i.get("language")

    
    data.append({"title": title})
    data.append({"language": language})

    with open("History_Books_Data.csv", "a") as f:
        f.write(f"{title}, {language}\n")

    


```
