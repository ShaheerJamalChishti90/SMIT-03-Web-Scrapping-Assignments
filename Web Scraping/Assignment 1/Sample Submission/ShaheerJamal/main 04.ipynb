```python
from bs4 import BeautifulSoup
import requests
import csv

url = "https://www.olx.com.pk/karachi_g4060695/jobs_c4"
web = requests.get(url).content
soup = BeautifulSoup(web, 'lxml')
```

```python
content = soup.find("div", {"class" : "_948d9e0a _371e9918 _95d4067f e1c7c3d4"})
content.prettify()
```

```python
# a = content.find_all("div", {"class" : "bee14b56"})
t = content.find_all("div", {"class" : "b8da62eb"})
l = content.find_all("div", {"class" : "c0c20674"})
d = content.find_all("div", {"class" : "c0c20674"})





for i, j, k in zip(t, l, d):
    title = i.find("h2", {"class" : "_941ffa5e"})
    location = j.find("span", {"class" : "_77000f35"})
    date = k.find("span", {"class" : "deca1dd3 d4fe84b4"}, {"aria-label":"Creation date"})
    # writer.writerow([Title, Location, Date])
    
    
    # print(f"Title: {title.text}")
    # print(f"Location: {location.text}")
    with open("OLX Jobs Data.csv", "a") as f:
        f.write(f"{title.text}, {location.text}, {date.text}\n")
        


```
